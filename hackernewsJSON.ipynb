```python
#Create a list that stores a certain number of items to work off of. We have the range of item ids from the MAX ID endpoint. I suggest that you start small and expand the dataset

baseurl = 'https://hacker-news.firebaseio.com/v0/item/'
new_list = []
for i in range(1,100):
    response1 = requests.get(baseurl + str(i) + '.json') 
    r = response1.json()
    new_list.append(r)
print(len(new_list))
pp(new_list)
```
