```python
import requests
import json
from pprint import pp
```

```python
baseurl = 'https://swapi.dev/api/starships/?page='
new_list = []

for i in range(1,5):
    response1 = requests.get(baseurl + str(i)) 
    r = response1.json()
    new_list.append(r)
#pp(new_list)

```

```python
result_list = []
model_names = []
cargo_list = []
crew_list = []
cost_list = []
films_list = []



for a in new_list:
    for b in a['results']:
        result_list.append(b)
print(result_list)

#problem1
def get_allnames():
    for c in result_list:
        model_names.append(c['name'])
    return model_names

print(get_allnames())


#problem2
def most_cargo():
    for d in result_list:
        if d['cargo_capacity'].isnumeric(): 
           cargo_list.append(int((d['cargo_capacity'])))

    for e in result_list:
       if e['cargo_capacity'].isnumeric() and int(e['cargo_capacity']) == max(cargo_list):
           print(e['name'])

print(most_cargo())

#problem3
def name_crew(shipname):
    for f in result_list:
        if f['starship_class'] == shipname:
            print('Name and crew of the {ship_name}:', (f['name'], f['crew']))

print(name_crew())


#problem4
def result_list():
    for g in result_list:
       if g['cost_in_credits'].isnumeric():
           cost_list.append(int(g['cost_in_credits']))
print(min(cost_list))

for h in result_list:
    if h['cost_in_credits'].isnumeric() and int(h['cost_in_credits']) == max(cost_list):
        print((h['cost_in_credits'], h['name'] ))

#problem5
cost = 100
for i in result_list:
    if i['cost_in_credits'].isnumeric() and int(i['cost_in_credits']) > cost:
         print(i['name'])


#problem6
film_list = []
x = 2
starship_list = []
for i in result_list:
     if len(i['films']) == x:
        starship_list.append(i['name'])
         
print('starship_list:', starship_list)
print('number of starships:', len(starship_list))

#problem7
length_list = []
def ShipMaxMinLength():
    for i in result_list:
        length_list.append(i['length'])
print(length_list)

        


        

         


```
